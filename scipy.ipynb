### #binomial distribution(no of success in n trials)

```python


#example
#e-commerce company:it is ovserved that 10% of thier customers returns the purchased items
#if 20 people bought products in one day them
```

```python


from scipy import stats
```

```python
#1.probability of no of returns is 5
#successes=5
#total=20
#success probality:0.1
stats.binom.pmf(5,20,0.1)
```

```python
#2.probability that a maximum of 5 customers will return
stats.binom.cdf(5,20,0.1)
```

```python
#3.probability that more than 5 will return
1-stats.binom.cdf(5,20,0.1)
```

```python
#4.average number of customers who likely to be return and variance
mean,var=stats.binom.stats(20,0.1)
print(mean,var)
```

```python
#visual representation of probability distribution 
import pandas as pd

data={'x':range(0,21),
     "pmf":stats.binom.pmf(range(0,21),20,0.1)
     }
```

```python
df=pd.DataFrame(data)
print(df)
```

```python
import seaborn as sns
```

```python
sns.barplot(x=df.x,y=df.pmf)
```

### #poison distribution(no of events occured in a period of time) 

```python
#example
#the no of calls occuring at a call center follows poison distribution at 10 callsper hour
```

```python
#1.probability that no of calls will be max 5
stats.poisson.cdf(5,10)
```

```python
#2.probality that no of calls over 3 hours will exceed 30
1-stats.poisson.cdf(30,30)
```

```python
#visual representation of probability distribution 
data={"x":range(20),
     "pmf":stats.poisson.pmf(range(20),10)}
```

```python
df=pd.DataFrame(data)
sns.barplot(x=df.x,y=df.pmf)
```
