# 2.Data Wrangling

## dealing missing data

```python


#replacing ? with NaN values
df1=df.replace('?',np.NaN)
 
```

```python
df1.info()
```

```python
df=df1.dropna(subset=['price'],axis=0)
```

```python
df.info()
```

```python
df['normalized-losses']
```

```python
df['normalized-losses']=df['normalized-losses'].replace(np.NaN,0)
df['normalized-losses']=df['normalized-losses'].astype('int64')
```

```python
df.dtypes
```

```python
mean=df['normalized-losses'].mean()
df['normalized-losses']=df['normalized-losses'].replace(0,mean)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['bore','stroke','horsepower','peak-rpm','num-of-doors'],axis=0,inplace=True)
```

```python
df.isnull().sum()
```

## normalization

```python
df[["length","width","height"]].head(4)
```

```python
#simple feature scaling
df1=df.copy(deep=True)
df1['length']=(df1["length"]/df1["length"].max())
```

```python
df1["width"]=df1["width"]/df["width"].max()
```

```python
df1["height"]=df1["height"]/df["height"].max()
```

```python
df1[["length","width","height"]].head(4)
```

```python
#min max 
df2=df.copy(deep=True)
```

```python
df2["length"]=(df2["length"]-df2["length"].min())/(df2["length"].max()-df2["length"].min())
```

```python
df2["width"]=(df2["width"]-df2["width"].min())/(df2["width"].max()-df2["width"].min())
```

```python
df2["height"]=(df2["height"]-df2["height"].min())/(df2["height"].max()-df2["height"].min())
```

```python
df2[["length","width","height"]].head(4)
```

```python
#z score
```

```python
df3=df.copy(deep=True)
```

```python
df3["length"]=(df3["length"]-df3["length"].mean())/df3["length"].std()
```

```python
df3["width"]=(df3["width"]-df3["width"].mean())/df3["width"].std()
```

```python
df3["height"]=(df3["height"]-df3["height"].mean())/df3["height"].std()
```

```python
df3[["length","width","height"]].head(4)
```

## binning


```python
bins=np.linspace(df1["price"].min(),df1["price"].max(),4)
```

```python
bins
```

```python
group_names=["low","middle","high"]
```

```python
df["price_binned"]=pd.cut(df["price"],bins=bins,labels=group_names,include_lowest=True)
```

```python
df.head(10)
```

```python
dummies=pd.get_dummies(df["fuel-type"])
```

```python
df=pd.concat([df,dummies],axis=1)
```

```python
df=df.drop(["fuel-type"],axis=1)
```
