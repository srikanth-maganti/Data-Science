```python
#Seaborn is a library for making statistical graphics in Python.
import seaborn as sns
```

```python
import pandas as pd
```

```python
data=pd.read_csv("Toyota.csv",na_values=[" ??"])
data.columns
```

```python
#Regression plots as the name suggests creates a regression line between 2 parameters and helps to visualize their linear relationships.
sns.set(style="darkgrid")
sns.regplot(data=data,x="Age",y="Price",marker="*")
```

```python
sns.lmplot(x="Age",y="Price",data=data,hue="FuelType")
```

```python
sns.countplot(data=data,x="FuelType")
```
