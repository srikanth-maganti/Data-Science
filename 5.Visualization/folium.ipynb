```python
import folium
```

```python
m1=folium.Map()
m1
```

```python
#for a specific loaction
m2=folium.Map(location=[41,29])
m2
```

```python
#By default, “tiles” appears as ”OpenStreetMap” in the Map() function.
#Stamen Toner, CartoDB positron, Cartodb dark_matter, Stamen Watercolor or Stamen Terrain.
```

```python
#tiles
m3=folium.Map(location=[41,29],width="50%",height="50%",tiles='CartoDB positron')
m3
```

```python
m3=folium.Map(location=[41,29],width="50%",height="50%",tiles='Cartodb dark_matter')
m3
```

```python
#specifying zoom 
m4=folium.Map(location=[41,29],width=500,height=500,zoom_start=10,max_zoom=12,min_zoom=8)
m4
```

```python
#
import folium
m5=folium.Map(location=[41,29],tiles="OpenStreetMap")
folium.CircleMarker(location=(41,29),radius=100,color='red',fill=True).add_to(m5)
m5
```

```python
m5=folium.Map(location=[41,29],tiles="OpenStreetMap")
folium.Circle(
radius=5000,
location=[41,29],
color='crimson',
fill=False,).add_to(m5)
m5
```

```python
m6=folium.Map(location=[41,29])
folium.Marker(location=[41,29],tooltip="click me",popup="<b>istanbul</b>").add_to(m6)
m6
```

```python
m=folium.Map(location=[41,29])
folium.TileLayer('Stamen Terrain').add_to(m)
folium.TileLayer('Stamen Toner').add_to(m)
folium.TileLayer('Stamen Water Color').add_to(m)
folium.TileLayer('cartodbpositron').add_to(m)
folium.TileLayer('cartodbdark_matter').add_to(m)
folium.LayerControl().add_to(m)
m
```
