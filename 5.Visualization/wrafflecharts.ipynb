```python
import pandas as pd 
from pywaffle import Waffle
import matplotlib.pyplot as plt
```

```python
value = [24, 23, 2, 15, 16]
labels=["a","b","c","d","e"]
plt.figure(
    FigureClass = Waffle,
    rows = 10,columns=10,
    values = value,labels=labels)

plt.show()
```

```python
value=[1,3,4,5,4]
plt.figure(FigureClass=Waffle,figsize=(2,10),values=value,rows=4,columns=5,title={"label":"maganti","loc":"left"})
plt.show()
```

```python
#using dictionary
dict={'DAYS':['MONDAY','TUESDAY','WEDNESDAY','THURSDAY','FRIDAY','SATURDAY','SUNDAY'],
      'FRUITS':['53','17','10','5','5','2','8'],
      'VEGETABLES':['25','35','12','13','5','4','6']
    }
df=pd.DataFrame.from_dict(dict)
df

```

```python
colours=['#8dd3c7','#bebada','#deb887','#80b1d3','#fb8072','#b3de69','#fdb462']

plot_labels=[f'{i} ({str(j)}%)' for i,j in zip(df.DAYS ,df.FRUITS)]

plot_labels
```

```python
# Convert 'FRUITS' column to numeric type
df['FRUITS'] = pd.to_numeric(df['FRUITS'])

# Drop rows with NaN values (if any)
df = df.dropna(subset=['FRUITS'])

# Convert the 'FRUITS' column to a list
values = df['FRUITS'].tolist()

# Create the waffle chart
plt.figure(
    FigureClass=Waffle,
    rows=5,
    columns=20,
    values=values,
    colors=colours,
    figsize=(10, 10),
    labels=plot_labels,
    legend={'loc': 'lower center', 'bbox_to_anchor': (0.5, -0.7), 'ncol': 3, 'fontsize': 12}
)
plt.show()
```

```python
df['FRUITS'] = pd.to_numeric(df['FRUITS'])


df = df.dropna(subset=['FRUITS'])

values = df['FRUITS'].tolist()


plt.figure(
    FigureClass=Waffle,
    rows=5,
    columns=20,
    values=values,
    colors=colours,
    figsize=(10, 10),
    labels=plot_labels,
    legend={'loc': 'lower center', 'bbox_to_anchor': (0.5, -0.7), 'ncol': 3, 'fontsize': 12},
    starting_location='SE'     #SE means SOUTH EAST,NW means NORTH WEST
)
plt.show()
```

```python
df['VEGETABLES'] = pd.to_numeric(df['VEGETABLES'])

plot_labels2=[f'{i} ({str(j)}%)' for i,j in zip(df.DAYS ,df.VEGETABLES)]


df = df.dropna(subset=['VEGETABLES'])

values = df['VEGETABLES'].tolist()


plt.figure(
    FigureClass=Waffle,
    rows=5,
    columns=20,
    values=values,
    colors=colours,
    figsize=(10, 10),
    labels=plot_labels2,
    legend={'loc': 'lower center', 'bbox_to_anchor': (0.5, -0.7), 'ncol': 3, 'fontsize': 12},
    #starting_location='SE'     
)
plt.show()
```

```python
plt.figure(
           FigureClass=Waffle,
           plots={211: {'values' :list(df["FRUITS"]),
                        'labels' :plot_labels,
                        'legend'  :{'loc': 'center left', 'bbox_to_anchor': (1.0, 0.5), 'ncol': 1, 'fontsize': 12},
                        'title'  :{'label':'Table1','fontsize':18}},
                  
                  212: {'values' :list(df["VEGETABLES"]),
                        'labels' :plot_labels2,
                        'legend'  :{'loc': 'center left', 'bbox_to_anchor': (1.0, 0.5), 'ncol': 1, 'fontsize': 12},
                        'title'  :{'label':'Table2','fontsize':18}}
                  
               
           },
    rows=5,
    columns=20,
    colors=colours,
    figsize=(20, 10)
    )
plt.show()
```

```python
off_colour='lightgrey'

fig,axs=plt.subplots(len(df),1,figsize=(10,15))

for (i,ax), color in zip(enumerate(axs.flatten()),colours):
    plot_colours=[color,off_colour]
    perc=df.iloc[i]['FRUITS']
    values=[perc,(100-perc)]
    day=df.iloc[i]['DAYS']
    Waffle.make_waffle(ax=ax,rows=5,columns=20,values=values,colors=plot_colours)
    ax.set_title(day)
    
plt.tight_layout()
plt.show()
```
